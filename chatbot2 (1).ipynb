```python
def chatbot():
    print(" Simple Chatbot (type 'bye' to exit)")

    while True:
        user_input = input("You: ").lower().strip()

        # Exit condition
        if user_input == "bye":
            print("Bot: Goodbye! ")
            break

        # Basic greetings
        elif user_input in ["hello", "hi"]:
            print("Bot: Hi there! ")

        elif user_input == "how are you":
            print("Bot: I'm fine, thanks!")

        # Math operations
        elif "add" in user_input or "+" in user_input:
            try:
                nums = [float(x) for x in user_input.replace("+", " ").split() if x.replace('.', '', 1).isdigit()]
                print("Bot: Result =", sum(nums))
            except:
                print("Bot: Please enter numbers properly!")

        elif "subtract" in user_input or "-" in user_input:
            try:
                nums = [float(x) for x in user_input.replace("-", " ").split() if x.replace('.', '', 1).isdigit()]
                if len(nums) >= 2:
                    result = nums[0]
                    for n in nums[1:]:
                        result -= n
                    print("Bot: Result =", result)
                else:
                    print("Bot: Need at least 2 numbers.")
            except:
                print("Bot: Error in subtraction.")

        elif "multiply" in user_input or "*" in user_input:
            try:
                nums = [float(x) for x in user_input.replace("*", " ").split() if x.replace('.', '', 1).isdigit()]
                result = 1
                for n in nums:
                    result *= n
                print("Bot: Result =", result)
            except:
                print("Bot: Error in multiplication.")

        elif "divide" in user_input or "/" in user_input:
            try:
                nums = [float(x) for x in user_input.replace("/", " ").split() if x.replace('.', '', 1).isdigit()]
                if len(nums) >= 2:
                    result = nums[0]
                    for n in nums[1:]:
                        result /= n
                    print("Bot: Result =", result)
                else:
                    print("Bot: Need at least 2 numbers.")
            except ZeroDivisionError:
                print("Bot: Cannot divide by zero!")
            except:
                print("Bot: Error in division.")

        # Default response
        else:
            print("Bot: I don't understand that yet.")

# Run chatbot
chatbot()
```
